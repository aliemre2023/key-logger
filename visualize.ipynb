```python
import matplotlib.pyplot as plt
```

```python
keys = dict()
```

```python
with open("data.txt", "r") as file:
    for line in file:
        values = line.split(',')
        key = values[0].strip()

        if key not in keys:
            keys[key] = 0
        keys[key] += 1
```

```python
keys = dict(sorted(keys.items(), key=lambda item: item[1], reverse=True))
print(keys)
```

```python
x_val = list(keys.keys())
y_val = list(keys.values())
plt.figure(figsize=(len(x_val) / 2, max(y_val) / 10))
plt.xlabel('Keys')
plt.ylabel('Times')
plt.title('Key Logger')
plt.xticks(rotation=45)

skills = ["'q'", "'w'", "'e'", "'r'"]
colors = ['blue' if key not in skills else 'red' for key in x_val]

plt.bar(x_val, y_val, color=colors)

plt.show()
```
